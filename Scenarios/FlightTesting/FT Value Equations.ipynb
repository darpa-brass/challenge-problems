```python
import math

import numpy as np
import matplotlib.pyplot as plt 
```

```python
# approximate value per kbps of throughput
def fun (x): 
    if x < 150:
        return 0.0
    elif x < 2000:
        return 100 - 100 * (math.e ** (-.005 * x))
    else:
        return 100 - 100 * (math.e ** (-.005 * 2000))
    

vfun = np.vectorize(fun)

x = np.linspace(0, 2500, 10000)
y = vfun(x)

plt.plot(x, y, '-')
plt.xlabel("Kbps Throughput")
plt.ylabel("Percentage Total Value")
plt.show()
```

```python
# approximate value per ms of latency
def fun (x): 
    if x < 50:
        return 100
    elif x < 60:
        ans = -(x-50)**2 + 100
        if (ans > 0):
            return ans
    else:
        return 0
    

vfun = np.vectorize(fun)

x = np.linspace(0, 125, 10000)
y = vfun(x)

plt.plot(x, y, '-')
plt.xlabel("Milliseconds of Latency")
plt.ylabel("Percentage Total Value")
plt.show()
```
